# We need to check our data

```python
import os
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
class ExploratoryDataAnalysis():
    
    def __init__(self,f_name,folder):
        self.f_name = pd.read_csv(f_name)
        self.folder = folder
        
    def draw_graph(self,label,count,title,xlab,ylab):
        plt.barh(count,label)
#         plt.xlabel(xlab)
#         plt.ylabel(ylab)
        plt.show()
    
    def getViz(self,col_name):
        dt = {}
        label = self.f_name[col_name].values
        for x in label:
            dt[x]=0
        for x in label:
            dt[x]+=1
        data=[]
        lab = []
        c=0
        for k,v in dt.items():
            lab.append(v)
            data.append(str(k))
            c+=1
        plt.barh(data,lab)
        plt.title("Count")
        plt.xlabel("Image Count")
        plt.ylabel("Labels")
        plt.show()

    def count_folder_image(self):
        label = []
        cnt = []
        cell_type_folder=os.listdir(self.folder)

        for x in cell_type_folder:
            cnt.append(len(os.listdir(self.folder+'/'+x)))
            label.append(x)
        
        plt.barh(label,cnt)
        plt.title("Count")
        plt.xlabel("Image Count")
        plt.ylabel("Labels")
        plt.show()
            
    
```

# From Above data we get Know that, it has some useless data and labels, Which can harm our data model and we need to remove it.

```python
if __name__ == "__main__":
    eda = ExploratoryDataAnalysis("data/train_images_labels.csv","data/images/TRAIN")
    eda.getViz("Category")
```
